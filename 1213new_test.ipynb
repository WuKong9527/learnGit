```python
import matplotlib.pyplot as plt
x = torch.linspace(-1, 1, steps=5)
y = torch.linspace(-1, 1, steps=5)

grid_x, grid_y = torch.meshgrid(x, y, indexing='xy')
print(grid_x)
print(grid_y)
z = grid_x**2 + grid_y**2
print(z)
# 绘制等高线图
plt.contourf(grid_x.numpy(), grid_y.numpy(), z.numpy(), levels=50, cmap='viridis')
plt.colorbar(label='z = x^2 + y^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot of z = x^2 + y^2')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import torch

# 定义 x 和 y 的取值范围
x = torch.linspace(-3, 3, steps=100)
y = torch.linspace(-3, 3, steps=100)

# 生成网格
grid_x, grid_y = torch.meshgrid(x, y, indexing='xy')

# 定义函数 z = sin(x^2 + y^2)
z = torch.sin(grid_x**2 + grid_y**2)

# 绘制 3D 曲面
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(grid_x.numpy(), grid_y.numpy(), z.numpy(), cmap='coolwarm')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z = sin(x^2 + y^2)')
plt.title('3D Surface Plot')
plt.show()
```

```python
# 定义空间范围
x = torch.linspace(-1, 1, steps=5)
y = torch.linspace(-1, 1, steps=5)
z = torch.linspace(-1, 1, steps=5)

# 生成 3D 网格
grid_x, grid_y, grid_z = torch.meshgrid(x, y, z, indexing='ij')

print(grid_x)
# 将网格点合并为点云
points = torch.stack([grid_x.flatten(), grid_y.flatten(), grid_z.flatten()], dim=-1)
print(points.shape)  # (125, 3)
print(points)  # 每行是一个点的 (x, y, z) 坐标
```

```python
# 图像大小 (5x5)
image = torch.arange(25).view(5, 5)

# 生成网格坐标
x = torch.linspace(0, 4, steps=5)  # 列坐标
y = torch.linspace(0, 4, steps=5)  # 行坐标
grid_x, grid_y = torch.meshgrid(x, y, indexing='xy')

# 使用网格采样
sampled_values = image[grid_y.long(), grid_x.long()]
print(image)
print(sampled_values)
```
